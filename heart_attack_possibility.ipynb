```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
scale.fit(hrt.drop('target',axis = 1))
```

```python
scaled_feat =scale.transform(hrt.drop('target',axis = 1))
```

```python
hrt_feat = pd.DataFrame(scaled_feat,columns = hrt.columns[:-1])
```

```python
hrt.columns
```

```python
hrt_feat.head()
```

```python
X = hrt_feat
y = hrt['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
pred = log.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(classification_report(y_test,pred))
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(accuracy_score(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predi = dtree.predict(X_test)
```

```python
predi
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(classification_report(y_test,predi))
print('\n')
print(confusion_matrix(y_test,predi))
print('\n')
print(accuracy_score(y_test,predi))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100)
```

```python
rfc.fit(X_train,y_train)
```

```python
predic = rfc.predict(X_test)
```

```python
predic
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,ConfusionMatrixDisplay
```

```python
print(classification_report(y_test,predic))
print('\n')
print(confusion_matrix(y_test,predic))
print('\n')
print(accuracy_score(y_test,predic))
```

```python
cons = confusion_matrix(y_test,predic , labels = rfc.classes_)
```

```python
consdisplay = ConfusionMatrixDisplay(confusion_matrix = cons, display_labels= ['0','1'])
```

```python
consdisplay.plot()
```
