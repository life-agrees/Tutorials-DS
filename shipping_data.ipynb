```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for cols in eco.columns:
    if eco[cols].dtype == 'object':
        eco[cols] = le.fit_transform(eco[cols])
```

```python
eco.head()
```

```python
eco.drop(['ID','Gender'],axis = 1,inplace = True)
```

```python
eco.head()
```

```python
reach_count = eco['Reached.on.Time_Y.N'].value_counts()
```

```python
plt.pie(reach_count.values,labels = reach_count.index,autopct = '%1.1f%%')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaled = StandardScaler()
```

```python
scaled.fit(eco.drop('Reached.on.Time_Y.N',axis = 1))
```

```python
scaled_feat = scaled.transform(eco.drop('Reached.on.Time_Y.N',axis = 1))
```

```python
eco_feat = pd.DataFrame(scaled_feat, columns = eco.columns[:-1])
```

```python
eco_feat.head()
```

```python
X = eco_feat
y = eco['Reached.on.Time_Y.N']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logm = LogisticRegression()
```

```python
logm.fit(X_train,y_train)
```

```python
predicts = logm.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predicts))
```

```python
print(confusion_matrix(y_test,predicts))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
pred = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100)
```

```python
rf.fit(X_train,y_train)
```

```python
prediction = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,prediction))
print('\n')
print(confusion_matrix(y_test,prediction))
```
